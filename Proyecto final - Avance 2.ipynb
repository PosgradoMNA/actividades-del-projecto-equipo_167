### **Escribe los pasos estándares en un máximo de 3 oraciones, por cada fase.**

**Fase I. Entendimiento del negocio**

Es la fase inicial que se enfoca en la comprensión de los objetivos del negocio en donde se toman decisiones y planteamos que objetivos queremos buscar en el proyecto con el análisis de los datos.

**Fase II. Entendimiento de los datos**

Consiste en el entendimiento o familiarización de los datos, es donde identificamos los problemas de calidad, descubrir conocimiento preliminar sobre los datos para empezar a formular hipótesis. 

**Fase III. Preparación**

Abarca todas las actividades para construir el conjunto final de datos tales como: selección de tablas, registros, transformaciones y limpieza de datos para la herramienta de modelado.

**Fase IV. Modelado**

Se realiza la selección y se aplican las técnicas de modelados que sea adecuadas al problema, y se calibran sus parámetros a valores óptimos.

**Fase V. Evaluación**

Consiste en la certificación de los modelos construidos, con el fin de garantizar que cumpla con los objetivos del negocio. Al final de esta fase, se debiera obtener una decisión sobre la aplicación de los resultados del análisis de datos.

**Fase VI. Despliegue**

Es la puesta en producción en donde el conocimiento obtenido tendrá que organizarse y presentarse para que el cliente pueda usarlo, con el fin de garantizar el desarrollo y la transformación del negocio.
